##MLOPS

## ‚úÖ What is MLOps?

**MLOps (Machine Learning Operations)** is a set of practices that combines **Machine Learning**, **DevOps**, and **Data Engineering** to streamline and automate the lifecycle of machine learning models ‚Äî from development to deployment and monitoring.

It aims to address the challenges associated with:

* Versioning and reproducibility
* Continuous training and deployment
* Monitoring model performance in production
* Scaling ML workflows across environments


## üß† Why MLOps is Needed

Traditional ML development is often experimental and isolated. MLOps is needed to:

* Bridge the gap between experimentation and production
* Ensure reproducibility of results
* Automate model training, testing, and deployment
* Enable continuous integration & delivery (CI/CD) for ML systems
* Monitor drift and retrain models when necessary
* Improve **collaboration between data scientists, ML engineers, and DevOps teams**

## üîÑ MLOps vs Traditional Machine Learning

| Aspect          | Traditional ML    | MLOps                            |
| --------------- | ----------------- | -------------------------------- |
| Focus           | Experimentation   | End-to-end automation            |
| Code Management | Often ad hoc      | Version-controlled               |
| Deployment      | Manual or one-off | Automated CI/CD pipelines        |
| Monitoring      | Rare              | Automated monitoring, retraining |
| Collaboration   | Silos             | Cross-functional teams           |
| Reproducibility | Inconsistent      | Enforced through tooling         |

## üîÅ Typical MLOps Flow (High-level)

1. **Data Ingestion & Processing**
2. **Model Training**
3. **Model Evaluation & Validation**
4. **Hyperparameter Tuning**
5. **Model Registry (Tracking)**
6. **Model Deployment**
7. **Monitoring & Retraining**

> Toolchains vary by cloud and organizational needs, but the principles remain consistent.

## üë• Who Works on MLOps (Roles by Stage)

| Stage                   | Role(s) Involved             |
| ----------------------- | ---------------------------- |
| Data ingestion/cleaning | Data Engineers               |
| Model development       | Data Scientists              |
| Pipeline automation     | ML Engineers                 |
| Deployment              | DevOps, ML Engineers         |
| Monitoring              | ML Engineers, SREs           |
| Governance & compliance | ML Engineers, Security Teams |


##Kubeflow

## üì¶ What is Kubeflow?

**Kubeflow** is an open-source ML platform built on Kubernetes that enables the orchestration of machine learning workflows.

It provides:

* **Pipelines**: for defining and managing complex workflows (train/test/tune/etc.)
* **Notebooks**: for collaborative development
* **Model Serving**: using KFServing or TensorFlow Serving
* **Metadata Tracking**, **Hyperparameter Tuning**, and more


## üéØ How Kubeflow Pipelines Work on Vertex AI (GCP)

When using Kubeflow Pipelines on **Vertex AI**, you benefit from GCP‚Äôs managed infrastructure:

1. **Define pipeline components**: Train, Test, Tune, etc., using Python functions decorated with `@dsl.component`.
2. **Assemble pipeline**: Connect components using `@dsl.pipeline`.
3. **Compile & Upload**: Compile the pipeline YAML and upload to Vertex AI.
4. **Run pipeline**: Trigger via Vertex AI Pipelines dashboard, SDK, or CI/CD.
5. **Track experiments**: Monitor via Vertex AI Experiments.

> Yes ‚Äî this is **part of MLOps**, specifically the **pipeline automation** stage (train/test/tune/deploy). Kubeflow helps enforce reproducibility and traceability.

## üì¶ Kubeflow Pipelines Structure

Kubeflow Pipelines is a platform for building and deploying portable, scalable ML workflows. Each pipeline is composed of multiple **components**, and these are orchestrated as a **DAG**.

---

## ‚úÖ Key Concepts & Structure

### üîπ Component

* A self-contained step in the ML process.
* Defined using `@component` or loaded from a YAML file.
* Accepts typed inputs and outputs.

**Syntax:**

```python
@component
def step_name(input1: Input[Dataset], output1: Output[Dataset], metrics: Output[Metrics]) -> Output:
```

---

### üîπ Pipeline

* Orchestrates multiple components.
* Uses `@pipeline` decorator.
* Defines the sequence and data flow.

**Syntax:**

```python
@pipeline(name='pipeline-name', description='...')
def pipeline_func(param1: str, param2: int):
    ...
```

---

### üîπ Input / Output Artifacts

Kubeflow uses `Input[...]` and `Output[...]` to manage artifacts between steps.

* `Input[Dataset]`, `Output[Dataset]`: Used to pass datasets between components.
* `Input[Model]`, `Output[Model]`: Used for model artifacts.
* `Output[Metrics]`: Used for evaluation metrics.

**Usage:**

* `.path` attribute is used to read/write files inside the container.
* Artifacts are automatically stored and versioned.

---

### üîπ Metrics

* Captures evaluation metrics like accuracy, precision, etc.
* Visualized in the Kubeflow/Vertex AI UI.
* Helps compare pipeline runs.

**Syntax:**

```python
metrics: Output[Metrics]
```

---

### üîπ NamedTuple

Used to return multiple values from a component, particularly for primitive outputs like paths, scores, or IDs.

````python
@component(packages_to_install=["google-cloud-aiplatform","gcsfs","imblearn","pandas","google-cloud-storage",'openpyxl'])
def validate_input_file(user_input_file:str)->NamedTuple("outputs", validation_result=bool):
    import pandas as pd
    df=pd.read_csv(user_input_file)
    if len(df.columns)==14 and set(df.columns)==set(['AGE','B','CHAS','CRIM','DIS','INDUS','LSTAT','NOX','PRICE','PTRATIO','RAD','RM','TAX','ZN']):
        validation_result=True
    else :
        validation_result=False
    outputs = NamedTuple("outputs", validation_result=bool)
    return outputs(validation_result)
````

Component 3 : model training

````python
@component(packages_to_install=["catboost","google-cloud-aiplatform","gcsfs","imblearn","pandas","google-cloud-storage",'openpyxl'])
def model_training(X_train_file:Input[Dataset], y_train_file:Input[Dataset],model_artifact: Output[Model]):

    import pickle
    from catboost import CatBoostRegressor

    with open(X_train_file.path, "rb") as file:
        x_train = pickle.load(file)

    with open(y_train_file.path, "rb") as file:
        y_train = pickle.load(file)

    cb=CatBoostRegressor()
    cb.fit(x_train,y_train)
    cb.save_model(model_artifact.path)
````

Component 4: Evaluation

````python
@component(packages_to_install=["google-cloud-aiplatform","gcsfs","imblearn","pandas","google-cloud-storage",'openpyxl','scikit-learn','catboost'])
def evaluation(X_test_file:Input[Dataset], y_test_file:Input[Dataset],model_artifacts :Input[Model],scalar_metrics: Output[Metrics]):
    from sklearn.metrics import mean_absolute_error,r2_score
    import pickle
    import pandas as pd
    from catboost import CatBoostRegressor
    cb=CatBoostRegressor()
    cb.load_model(model_artifacts.path)

    with open(X_test_file.path, "rb") as file:
        X_test = pickle.load(file)

    with open(y_test_file.path, "rb") as file:
        y_test = pickle.load(file)

    predictions = cb.predict(X_test)
    Testing_accuracy=cb.score(X_test,y_test)
    mean_absolute_error=mean_absolute_error(y_test,predictions)
    r2=r2_score(y_test,predictions)
    scalar_metrics.log_metric("Testing_accuracy", Testing_accuracy)
    scalar_metrics.log_metric("mean_absolute_error", mean_absolute_error)
    scalar_metrics.log_metric("r2", r2)
````

## üîç Compiler and Compiler Parameters

The **Kubeflow compiler** transforms your pipeline Python function into a JSON/YAML file deployable on Vertex AI.

**Syntax:**

```python
compiler.Compiler().compile(
    pipeline_func=pipeline_func,
    package_path="path/to/pipeline.json"
)
```

### `package_path` (in `compile`)

* Location where the compiled pipeline specification (JSON/YAML) is saved.
* Required for submitting to Vertex AI.

---

## üöÄ Vertex AI Pipeline Job Parameters

**PipelineJob** is used to launch the compiled pipeline on Google Cloud Vertex AI.

**Syntax:**

```python
aiplatform.PipelineJob(
    display_name="job-name",
    template_path="path/to/pipeline.json",
    pipeline_root="gs://your-bucket/pipeline-root",
    parameter_values={"param1": "value1"}
)
```

### `template_path`

* Path to the compiled pipeline specification file (e.g., from `package_path` above).
* Can be local or on GCS.

### `pipeline_root`

* GCS bucket path where all artifacts (input/output data, models, metrics) are stored during the pipeline run.
* Essential for reproducibility, traceability, and lineage.

---

## üß† Experiments, Metadata & Tracking in Vertex AI

* **Experiment**: A logical grouping of pipeline runs, useful for organizing experiments (e.g., different hyperparameter configs).
* **Run**: A single pipeline execution.
* **Step/Task**: Individual component execution.
* **Artifacts**: Outputs like datasets, metrics, models.

````python
from kfp.v2 import dsl, compiler
from google.cloud import aiplatform
import uuid

# Generate a unique experiment name
id = str(uuid.uuid4())[:8]
EXPERIMENT_NAME = f"experiment-hyperparameters-catboostregressor-project-{id}"

# Define hyperparameter runs
runs = [
    {"iterations": 500, "learning_rate": 0.1, "depth": 4, "l2_leaf_reg": 2, "bagging_temperature": 2},
    {"iterations": 1000, "learning_rate": 0.03, "depth": 6, "l2_leaf_reg": 3, "bagging_temperature": 1},
]

# Compile the pipeline
compiler.Compiler().compile(
    pipeline_func=CATBoostRegressor_Project001_pipeline,  # Replace with your pipeline function
    package_path="CATBoostRegressor_Project001_experiment_HP.json"
)

# Initialize Vertex AI
aiplatform.init(project="PROJECT, location="europe-west2")

# Submit runs
for i, run in enumerate(runs):
    job = aiplatform.PipelineJob(
        display_name=f"catboost-hyperparam-run-{i+1}",
        template_path="CATBoostRegressor_Project001_experiment_HP.json",
        pipeline_root="gs://data-science-projects-dev/AI_SOFTWARE_TESTING/TRA_Kubeflow_pipelines/Experiment-HP-CATBoostRegressorProject001",
        parameter_values={
            "csv_path": 'gs://data-science-projects-dev/AI_SOFTWARE_TESTING/TRA_Kubeflow_pipelines/CAT-Boost-Regressor-Prooject001/Boston_house_price.csv',
            **run
        },

        enable_caching=False,
        job_id=f"catboost-proj-{id}-{i+1}"
    )
    job.submit(experiment=EXPERIMENT_NAME)
````
